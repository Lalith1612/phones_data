Detection of outliers using z-zscore method

```python
from scipy import stats
```

```python
z=stats.zscore(data["highest_price"])
z
```

Positions of z-score

```python
as1=np.where(z>0)
as1
```

Detection of outliers using IQR method

```python
Q1=np.percentile(data["highest_price"],25,interpolation='midpoint')
Q3=np.percentile(data["highest_price"],75,interpolation='midpoint')
IQR=Q3-Q1
print(Q1)
print(Q3)
print(IQR)
print("shape",data.shape)
```

```python
upper=data["highest_price"]>=(Q3+1.5*IQR)
print("upper: ",upper)
print(np.where(upper))
lower=data["highest_price"]<=(Q3-1.5*IQR)
print("lower: ",lower)
print(np.where(lower))
print("new shape",data.shape)
```

```python
mu = 1000
std = 5000
x = np.linspace(start=1, stop=10, num=100)
y = stats.norm.pdf(x, mu, std) 
plt.plot(x, y, label='phones')
plt.xlabel('brand_name', fontsize=15)
plt.ylabel('battery_size', fontsize=15)
plt.grid(b=True, color='DarkTurquoise', alpha=0.2, linestyle=':', linewidth=2)
plt.rcParams['figure.figsize'] = [10/2.54, 8/2.54]
plt.legend()
plt.show()
```
